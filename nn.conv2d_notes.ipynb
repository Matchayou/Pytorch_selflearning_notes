```python
import torch 
import torchvision
from torch.utils.data import DataLoader
```

```python
from torch import nn
from torch.nn import Conv2d
from torch.utils.tensorboard import SummaryWriter


dataset = torchvision.datasets.CIFAR10(
    root='./dataset',train=False,download=True,#.的多少代表目录的层级
    transform=torchvision.transforms.ToTensor())
dataloader = DataLoader(dataset, batch_size=64)

class Tudui(nn.Module):
    def __init__(self):
        super(Tudui, self).__init__()#调用父类 nn.Module 的初始化方法
        self.conv1 = Conv2d(in_channels=3, out_channels=6, kernel_size=3, stride=1, padding=0)

    def forward(self, x):
        #forward 方法定义了数据流经网络时的操作。
        # 在这里，输入 x 通过卷积层 self.conv1(x)，然后返回卷积后的结果。
        x = self.conv1(x)
        return x
    
tudui = Tudui()
print(tudui)#会输出模型的结构

writer = SummaryWriter("../tudui")
step = 0
for data in dataloader:
    imgs, targets = data
    output = tudui(imgs)
    print(imgs.shape)  # 输入形状
    print(output.shape)  # 输出形状
    writer.add_images("input", imgs, step)
    #torch.Size([64, 6, 30, 30])因为维度过多，无法被add。images处理，所以需要reshape
    #torch.Size([64, 6, 30, 30])→ [xxx,3,30,30]
    #相当于把多层的数据平铺开来
    output = torch.reshape(output, (-1, 3, 30, 30))
    writer.add_images("output", output, step)

    step += 1
```
